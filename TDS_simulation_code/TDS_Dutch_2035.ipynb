### Frequency & RoCoF response Plots

```python
models_raw = {
    "No FFR": ("MODEL_A", "ss_A"),
    "Droop": ("MODEL_B", "ss_B"),
    "Decentralised MPC": ("MODEL_C", "ss_C"),
    "Centralised MPC": ("MODEL_D", "ss_D"),
    "Distributed MPC 3": ("MODEL_E", "ss_E"),
    "Distributed MPC 8": ("MODEL_G", "ss_G"),
    "Robust MPC": ("MODEL_F", "ss_F")
}

colors = {
    "No FFR": "#999999",
    "Droop": "#D55E00",
    "Decentralised MPC": "#0072B2",
    "Centralised MPC": "#CC79A7",
    "Distributed MPC 3": "#009E73",
    "Distributed MPC 8": "#7BE6C3",
    "Robust MPC": "#E69F00"
}

enabled_models = {
    "No FFR": False,
    "Droop": False,
    "Decentralised MPC": True,
    "Centralised MPC": True,
    "Distributed MPC 3": True,
    "Distributed MPC 8": True,
    "Robust MPC": True,
}

plot_background_signals = False   # False = only COI curves
background_alpha = 0.7
background_linewidth = 1

freq_ylim = (49.9, 50.02)
rocof_ylim = (-0.1, 0.05)

def lighten_color(color, amount=0.30):
    """
    Blend color with white.
    amount=0 -> original color
    amount=1 -> white
    """
    c = np.array(to_rgb(color))
    white = np.array([1.0, 1.0, 1.0])
    return tuple((1 - amount) * c + amount * white)

def set_nice_yticks(ax, n_ticks=5, fixed_ylim=None):
    if fixed_ylim is not None:
        ymin, ymax = fixed_ylim
        ax.set_ylim(ymin, ymax)

        ticks = np.linspace(ymin, ymax, n_ticks)
        ax.set_yticks(ticks)
        return

    ymin, ymax = ax.get_ylim()
    span = ymax - ymin
    if span == 0:
        return

    raw_step = span / (n_ticks - 1)

    exp = np.floor(np.log10(raw_step))
    frac = raw_step / (10 ** exp)

    if frac <= 1:
        nice_frac = 1
    elif frac <= 2:
        nice_frac = 2
    elif frac <= 2.5:
        nice_frac = 2.5
    elif frac <= 5:
        nice_frac = 5
    else:
        nice_frac = 10

    step = nice_frac * (10 ** exp)
    nice_ymin = np.floor(ymin / step) * step
    nice_ymax = np.ceil(ymax / step) * step
    ticks = np.arange(nice_ymin, nice_ymax + 0.5 * step, step)

    ax.set_ylim(nice_ymin, nice_ymax)
    ax.set_yticks(ticks)

def resolve(ref):
    if not isinstance(ref, str):
        return ref
    val = globals().get(ref)
    if val is not None:
        return val
    if os.path.exists(ref):
        return ref
    return None

# -----------------------------
# Collect results for enabled and resolvable models only
# -----------------------------
results = {}

for name, (model_ref, ss_ref) in models_raw.items():
    if not enabled_models.get(name, False):
        continue

    model = resolve(model_ref)
    ss = resolve(ss_ref)

    if model is None:
        print(f"Skipping '{name}': model reference {model_ref!r} was not found/resolved.")
        continue
    if ss is None:
        print(f"Skipping '{name}': simulation object {ss_ref!r} was not found/resolved.")
        continue

    try:
        gen_df = pd.read_excel(model, sheet_name="GENROU")
    except Exception as e:
        print(f"Skipping '{name}': could not read GENROU from {model!r}: {e}")
        continue

    try:
        df = ss.dae.ts.df
    except Exception as e:
        print(f"Skipping '{name}': could not access ss.dae.ts.df on {ss!r}: {e}")
        continue

    time = df.index

    f_df = df.filter(regex=r"^f BusROCOF")
    r_df = df.filter(regex=r"^Wf_y BusROCOF")

    if f_df.shape[1] == 0:
        print(f"Skipping '{name}': no f BusROCOF columns found.")
        continue

    bus_numbers = f_df.columns.str.extract(r"(\d+)").astype(float)[0]

    try:
        gen_df["weight"] = gen_df["Sn"] * (gen_df["M"] / 2)
        bus_weights = gen_df.groupby("bus")["weight"].sum()
    except Exception as e:
        print(f"Skipping '{name}': error computing generator weights: {e}")
        continue

    bus_numbers_filled = bus_numbers.fillna(-1).astype(int)
    weights = bus_numbers_filled.map(bus_weights).fillna(0.0).to_numpy()

    if weights.sum() == 0:
        print(f"Skipping '{name}': computed weights sum to zero.")
        continue

    f_coi = (f_df.to_numpy() @ weights) / weights.sum()
    r_coi = (r_df.to_numpy() @ weights) / weights.sum()

    results[name] = {
        "time": time,
        "f_coi": f_coi,
        "r_coi": r_coi,
        "f_df": f_df,
        "r_df": r_df,
    }


# -----------------------------
# If no results, exit gracefully
# -----------------------------
if not results:
    print("No valid models were processed. Nothing to plot.")
else:
    fig, axes = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

    plt.rcParams.update({
        "font.size": 18,
        "axes.titlesize": 18,
        "axes.labelsize": 18,
        "xtick.labelsize": 18,
        "ytick.labelsize": 18,
        "legend.fontsize": 18,
        "figure.titlesize": 18
    })

    for name, d in results.items():
        color = colors.get(name, None)
        bg_color = lighten_color(color, amount=0.3) if color is not None else color

        time = d["time"]

        if plot_background_signals:
            for col in d["f_df"].columns:
                axes[0].plot(
                    time,
                    50 * d["f_df"][col],
                    color=bg_color,
                    alpha=background_alpha,
                    linewidth=background_linewidth,
                    zorder=1,
                    label="_nolegend_"
                )

        axes[0].plot(
            time,
            50 * d["f_coi"],
            label=name,
            color=color,
            linewidth=2.0,
            zorder=3
        )

    axes[0].set_ylabel("COI Frequency [Hz]")
    axes[0].set_title("Frequency response")
    axes[0].grid(True)
    axes[0].legend(loc="upper right")

    set_nice_yticks(axes[0], n_ticks=5, fixed_ylim=freq_ylim)
    set_nice_yticks(axes[1], n_ticks=5, fixed_ylim=rocof_ylim)
    for name, d in results.items():
        color = colors.get(name, None)
        bg_color = lighten_color(color, amount=0.3) if color is not None else color

        time = d["time"]

        if plot_background_signals:
            for col in d["r_df"].columns:
                axes[1].plot(
                    time,
                    50 * d["r_df"][col],
                    color=bg_color,
                    alpha=background_alpha,
                    linewidth=background_linewidth,
                    zorder=1,
                    label="_nolegend_"
                )

        axes[1].plot(
            time,
            50 * d["r_coi"],
            label=name,
            color=color,
            linewidth=2.0,
            zorder=3
        )

    axes[1].set_xlabel("Time [s]")
    axes[1].set_ylabel("COI RoCoF [Hz/s]")
    axes[1].set_title("RoCoF response")
    axes[1].grid(True)

    axes[0].margins(x=0, y=0.05)
    axes[1].margins(x=0, y=0.05)

    set_nice_yticks(axes[0], n_ticks=5, fixed_ylim=freq_ylim)
    set_nice_yticks(axes[1], n_ticks=5, fixed_ylim=rocof_ylim)

    plt.tight_layout()

    def sanitize(s):
        return re.sub(r"\W+", "", s)

    plt.show()
```

### Active power injections Plots

```python
inverters = [43, 44, 52, 53, 54, 55, 56, 57]

# Keep your models dict but allow ss entries to be either an object or a string name (safe if undefined)
models_raw = {
    "Droop": {
        "ss": "ss_B",
        "regex": r"^Pe REGF1W",
        "col": "Pe REGF1W {}",
        "color": "#D55E00"
    },
    "Decentralised MPC": {
        "ss": "ss_C",
        "regex": r"^Pe REGF1W MPC DEC",
        "col": "Pe REGF1W MPC DEC {}",
        "color": "#0072B2"
    },
    "Centralised MPC": {
        "ss": "ss_D",
        "regex": r"^Pe REGF1W MPC CEN",
        "col": "Pe REGF1W MPC CEN {}",
        "color": "#CC79A7"
    },
    "Distributed MPC 3": {
        "ss": "ss_E",
        "regex": r"^Pe REGF1W MPC DIS",
        "col": "Pe REGF1W MPC DIS {}",
        "color": "#009E73"
    },
    "Distributed MPC 8": {
        "ss": "ss_G",
        "regex": r"^Pe REGF1W MPC DIS",
        "col": "Pe REGF1W MPC DIS {}",
        "color": "#7BE6C3"
    },
    "Robust MPC": {
        "ss": "ss_F",
        "regex": r"^Pe REGF1W MPC STA",
        "col": "Pe REGF1W MPC STA {}",
        "color": "#E69F00"
    }
}

enabled_models = {
    "Droop": False,
    "Decentralised MPC": True,
    "Centralised MPC": True,
    "Distributed MPC 3": True,
    "Distributed MPC 8": True,
    "Robust MPC": True
}

inverter_names = {
    43: "Vattenfall I&II",
    44: "Vattenfall III&IV",
    52: "Beta Zeevonk II",
    53: "Alpha Noordzeker",
    54: "Gamma A&B",
    55: "Nederwiek 1",
    56: "Nederwiek 2",
    57: "Doordewind"
}

def set_nice_yticks(ax, n_ticks=5):
    ymin, ymax = ax.get_ylim()
    span = ymax - ymin
    if span == 0:
        return

    # target tick spacing
    raw_step = span / (n_ticks - 1)

    # round step to a "nice" number: 1, 2, 2.5, 5, 10 × power of 10
    exp = np.floor(np.log10(raw_step))
    frac = raw_step / (10 ** exp)

    if frac <= 1:
        nice_frac = 1
    elif frac <= 2:
        nice_frac = 2
    elif frac <= 2.5:
        nice_frac = 2.5
    elif frac <= 5:
        nice_frac = 5
    else:
        nice_frac = 10

    step = nice_frac * (10 ** exp)

    # snap limits to multiples of the step
    nice_ymin = np.floor(ymin / step) * step
    nice_ymax = np.ceil(ymax / step) * step

    ticks = np.arange(nice_ymin, nice_ymax + 0.5 * step, step)

    ax.set_ylim(nice_ymin, nice_ymax)
    ax.set_yticks(ticks)
# -----------------------------
# Helper: resolve references that may be:
# - actual object
# - name of a variable in globals() (string)
# Returns resolved object or None
# -----------------------------
def resolve(ref):
    if not isinstance(ref, str):
        return ref
    val = globals().get(ref)
    if val is not None:
        return val
    # not found
    return None

# -----------------------------
# Pre-extract dataframes for only enabled & resolvable models
# -----------------------------
data = {}
for name, m in models_raw.items():
    if not enabled_models.get(name, False):
        continue

    ss_ref = m.get("ss")
    ss = resolve(ss_ref)
    if ss is None:
        print(f"Skipping '{name}': simulation object {ss_ref!r} not found.")
        continue

    # try to access the timeseries DataFrame
    try:
        df = ss.dae.ts.df
    except Exception as e:
        print(f"Skipping '{name}': cannot access ss.dae.ts.df -> {e}")
        continue

    P_df = df.filter(regex=m["regex"])
    time = df.index

    # store only if DataFrame object created successfully (may be empty columns)
    data[name] = {
        "P_df": P_df,
        "time": time,
        "col": m["col"],
        "color": m["color"]
    }

# -----------------------------
# Plotting (modified)
# -----------------------------
fig, axes = plt.subplots(4, 2, figsize=(14, 16), sharex=True)
axes = axes.flatten()

plt.rcParams.update({
    "font.size": 18,          # base size (affects most text)
    "axes.titlesize": 18,     # subplot titles
    "axes.labelsize": 18,     # x/y axis labels
    "xtick.labelsize": 18,    # x tick labels
    "ytick.labelsize": 18,    # y tick labels
    "legend.fontsize": 18,    # legend text
    "figure.titlesize": 18    # suptitle
})

# Track which models actually produced plotted traces (used for filename)
plotted_models = []
# Collect legend handles/labels globally (deduplicate preserving order)
legend_map = {}

# NEW: track whether we've already added the global "Power limit" legend entry
power_limit_added = False

for ax, inv in zip(axes, inverters):
    limit_reference = None
    axis_plotted = False
    for name, d in data.items():
        col = d["col"].format(inv)

        if col not in d["P_df"].columns:
            continue

        P = 2000 * d["P_df"][col]
        if P.empty:
            continue

        if limit_reference is None:
            limit_reference = P.iloc[0]

        # plot model trace
        ln, = ax.plot(
            d["time"],
            P,
            color=d["color"],
            label=name
        )
        axis_plotted = True

        # record handle for global legend (deduplicate)
        if name not in legend_map:
            legend_map[name] = ln
        if name not in plotted_models:
            plotted_models.append(name)

    # Limit line (add single legend entry for the first dashed line)
    if limit_reference is not None:
        t_fin = globals().get("t_fin", None)
        if t_fin is None:
            last_times = [d["time"] for d in data.values() if hasattr(d["time"], 'max')]
            if last_times:
                try:
                    t_fin = max([t.max() for t in last_times])
                except Exception:
                    t_fin = None
        if t_fin is None:
            try:
                xlims = ax.get_xlim()
                t_fin = xlims[1]
            except Exception:
                t_fin = 0

        lim = limit_reference / 0.9

        # If we haven't added the "Power limit" legend handle yet, plot with a label and store handle
        if not power_limit_added:
            lim_ln, = ax.plot([0, t_fin], [lim, lim],
                               color='grey', linestyle='dashed',
                               label="Power limit")
            # add to global legend map once
            legend_map["Power limit"] = lim_ln
            power_limit_added = True
        else:
            # draw the same line on this axis but don't label it again
            ax.plot([0, t_fin], [lim, lim], color='grey', linestyle='dashed')
   
    name = inverter_names.get(inv, f"Inverter {inv}")
    ax.set_title(name, fontsize=22)
    ax.grid(True)

for ax in axes:
        ax.margins(x=0, y=0.05)
        set_nice_yticks(ax, n_ticks=5)

# Common labels
fig.supxlabel("Time [s]")
fig.supylabel("Active Power [MW]")
#fig.suptitle("FFR Active Power Response", fontsize=14)

# Global legend: use handles from legend_map in insertion order
if legend_map:
    handles = [legend_map[k] for k in legend_map.keys()]
    labels = list(legend_map.keys())
    axes[1].legend(handles, labels, loc="upper right")
else:
    print("Warning: no model traces were plotted; legend will be empty.")

plt.tight_layout(rect=[0, 0, 0.95, 0.96])

# -----------------------------
# Filename generation & save
# -----------------------------
def sanitize(s):
    return re.sub(r"\W+", "", s)

def fmt_val(v):
    if v is None:
        return "NA"
    s = str(v)
    s = s.replace(".", "p")
    return re.sub(r"\W+", "", s)

plt.show()
```
